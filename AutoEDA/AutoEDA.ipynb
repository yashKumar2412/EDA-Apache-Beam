```python
import pandas as pd
df = pd.read_csv('heart_disease_uci.csv')
```

```python
df.head()
```

```python
import sweetviz as sv
my_report = sv.analyze(df)
my_report.show_html('eda_report.html')
```

```python
age_report = sv.analyze([df, "Heart Disease Dataset"], target_feat='age')
age_report.show_html('age_analysis_report.html')
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(df, test_size=0.2)

report = sv.compare([train_data, "Train"], [test_data, "Test"])
report.show_html('comparative_report.html')
```
